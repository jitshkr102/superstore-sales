```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("train.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%d/%m/%Y')
```

```python
df['Ship Date'] = pd.to_datetime(df['Ship Date'], format='%d/%m/%Y')
```

```python
df['Month_Num'] = df['Order Date'].dt.month
```

```python
df['Month'] = df['Order Date'].dt.strftime('%b')
```

```python
df['Month'].head()
```

```python
df.head()
```

```python
df.drop('Row ID', axis = 1)
```

```python
df['Year'] = df['Order Date'].dt.year
```

```python
df.head()
```

```python
yearly_sales = df.groupby('Year')['Sales'].sum()
```

```python
yearly_sales
```

```python
yearly_sales.plot(kind = 'bar', color = 'skyblue', figsize = (12,4))
plt.title("Yearly total sales")
plt.xlabel("Year")
plt.ylabel("Sales")
plt.show()

```

## 📊 Insight: Year-wise Sales Analysis

- 2018 recorded the highest sales among all years
- This indicates strong business performance in that year
- Sales may be influenced by expansion, promotions, or market demand

```python
monthly_sales = df.groupby(['Month', 'Month_Num'])['Sales'].sum().reset_index()
monthly_sales = monthly_sales.sort_values(by= 'Month_Num')
```

```python
#monthly_sales.set_index('Month', inplace=True)
```

```python
plt.plot(monthly_sales['Month'], monthly_sales['Sales'], marker='o')

plt.title("Monthly Total Sales")
plt.xlabel("Months")
plt.ylabel("Sales")

plt.show()
```

```python
df.head()
```

```python
df.columns
```

```python
category_sales = df.groupby('Category')['Sales'].sum()
category_sales
```

```python
sub_category = df.groupby('Sub-Category')['Sales'].sum()
sub_category = sub_category.sort_values(ascending= False)
sub_category
```

```python
sub_category.plot(kind = 'bar', color = 'green', figsize = (8,3))
plt.title("Items with Sales")
plt.xlabel("Category")
plt.ylabel("Sales")
plt.show()
```

```python
region_sales = df.groupby('Region')['Sales'].sum()
region_sales = region_sales.sort_values(ascending= False)
region_sales
```

```python
region_sales.plot(kind = 'bar', color = 'red' , figsize = (3,3))
plt.title("Regionwith sales")
plt.xlabel("Region")
plt.ylabel("Sales")
plt.show()
```

```python
df.columns
```

```python
df['Segment']
```

```python
segment_sales = df.groupby('Segment')['Sales'].sum()
segment_sales = segment_sales.sort_values()
segment_sales = segment_sales.astype(int)
```

```python
segment_sales.plot(kind = 'bar', color = 'green', figsize = (3,3))
plt.xlabel('Segment')
plt.ylabel('Sales')
plt.show()
```

```python
pivot_data = df.pivot_table(
    values='Sales',
    index='Region',
    columns='Segment',
    aggfunc='sum'
)
```

```python
pivot_data.plot(kind = 'bar',stacked = True, figsize = (8,6))
plt.title('Region-wise Sales by Segment')
plt.xlabel('Regions')
plt.ylabel('Sales')
plt.show()
```

```python
pivot_data_category = df.pivot_table(
    values = 'Sales',
    index = 'Segment',
    columns = 'Category',
    aggfunc = 'sum'
)
```

```python
pivot_data_category.plot(kind = 'bar', stacked = True)
plt.xlabel('Category')
plt.ylabel('Sales')
plt.show()
```
